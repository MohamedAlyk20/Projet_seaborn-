## Seaborn est basé sur Matplotlib mais orienté vers les analyses statistiques visuelles.

## je vais devoir renommer les titres de mon data set

```python
df.rename(columns={'name':'nom_de_voiture',
                   'year':'annee',
                   'selling_price':'prix_de_vente',
                   'km_driven':'km_parcourus',
                   'fuel':'carburant',
                   'seller_type':'type_de_vendeur',
                   'transmission':'boite_de_vitesse', 
                   'owner':'proprietaire_precedent',
                    },inplace=True)
```

## 1. Repartition des prix de vente des voitures 

```python
sns.histplot(data=df, x='prix_de_vente',bins=100, kde=True, color='orange')
plt.title("1. Distribution des prix de vente")
plt.xlabel("Prix de vente")
plt.ylabel("Nombre de voitures")
plt.show()
```

## 2. Prix de vente de fonction du kilometrage

```python
sns.scatterplot(data=df, x='km_parcourus', y='prix_de_vente',color="gold", alpha=0.6)
plt.title("2. Prix en fonction du Kilométrage")
plt.xlabel("Kilomètres parcourus")
plt.ylabel("Prix de vente")
plt.show()
```

## 3. Prix de vente Selon le type de carburant 

```python
sns.boxplot(data=df, x='carburant', y='prix_de_vente', color='gold')
plt.title("3. Prix selon le carburant")
plt.xlabel("Carburant")
plt.ylabel("Prix de vente")
plt.show()
```

##  4. Prix selon l’année du modèle

```python
sns.scatterplot(data=df, x='annee', y='prix_de_vente')
plt.title("4. Prix selon l'année")
plt.xlabel("Année")
plt.ylabel("Prix de vente")
plt.show()
```

##  5. Prix selon le type de vendeur

```python
sns.boxplot(data=df, x='type_de_vendeur', y='prix_de_vente')
plt.title("5. Prix selon le type de vendeur")
plt.xlabel("Type de vendeur")
plt.ylabel("Prix de vente")
plt.show()
```

## 6. Prix selon la boîte de vitesse

```python
sns.boxplot(data=df, x='boite_de_vitesse', y='prix_de_vente')
plt.title("6. Prix selon la boîte de vitesse")
plt.xlabel("Boîte de vitesse")
plt.ylabel("Prix de vente")
plt.show()
```

## 7. Prix selon le nombre de propriétaires précédents

```python
sns.boxplot(data=df, x='proprietaire_precedent', y='prix_de_vente')
plt.title("7. Prix selon les propriétaires précédents")
plt.xlabel("Propriétaire précédent")
plt.ylabel("Prix de vente")
plt.xticks(rotation=45)
plt.show()
```

## 8. Corrélation entre les variables numériques

```python
sns.heatmap(df[['annee', 'prix_de_vente', 'km_parcourus']].corr(), annot=True, cmap='coolwarm')
plt.title("8. Corrélation entre variables numériques")
plt.show()
```

## 9. Pairplot des variables principales

```python
df_clean = df[['annee', 'prix_de_vente', 'km_parcourus', 'carburant']].dropna()
df_clean = df_clean[df_clean['carburant'].apply(lambda x: isinstance(x, str))]

sns.pairplot(df_clean, hue='carburant')
plt.suptitle("9. Analyse croisée des variables", y=1.02)
plt.show()
```

## 10. Marques les plus fréquentes

```python
df['marque'] = df['nom_de_voiture'].apply(lambda x: x.split()[0])
top_marques = df['marque'].value_counts().nlargest(10).index

plt.figure(figsize=(10, 4))
sns.countplot(data=df[df['marque'].isin(top_marques)], x='marque', order=top_marques)
plt.title("10. Marques les plus fréquentes")
plt.xlabel("Marque")
plt.ylabel("Nombre de voitures")
plt.show()
```
